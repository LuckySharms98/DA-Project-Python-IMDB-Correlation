```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 

##what's the correlation between budget and gross?
##what fields directly correlate with gross revenue?
```

```python
mvs = pd.read_csv('movies.csv')
mvs.head()
```

```python
##Check for empty cells
for col in mvs.columns:
    percent_missing = np.mean([mvs[col].isnull()])
    print(f'{col} - {percent_missing}%')
```

```python
#Let's drop all rows with missing data for the time being:
mvs = mvs.dropna()
for col in mvs.columns:
    percent_missing = np.mean([mvs[col].isnull()])
    print(f'{col} - {percent_missing}%')
```

```python
#look at field types
mvs.dtypes
#we should convert budget and gross to int64 instead of float
mvs.budget = mvs['budget'].astype('int64')
mvs.gross = mvs['gross'].astype('int64')
mvs.dtypes
```

```python
#Next, we need to account for the fact that the 'release' values differ than those in 'year'. 
#This will be a problem with our analysis. Let's extract the year string from 'released' in a new col.
mvs['year_corr'] = mvs.released.str.extract(pat = '([0-9]{4})').astype(int)
mvs.head()
```

```python
## now lets compare year and year_corr to look for differences
#mvs['test'] = [x for x in if mvs['year'] == mvs['year_corr']]
```

```python
#Sort by gross
mvs.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#drop all duplicates
mvs.drop_duplicates()
```

```python
#what's most correlated to gross revenue?
#budget vs. gross revenue - Scatter plot
plt.scatter(x=mvs.budget,y=mvs.gross)
plt.xlabel('Gross Revenue')
plt.ylabel('Film Budget')
plt.title('Film Budget vs. Gross Revenue')
```

```python
##regression with above plot (seaborn)
sns.regplot(x='budget',y='gross',data = mvs, scatter_kws = {"color":"red"}, line_kws={"color":"blue"})
```

```python
##what exactly is this correlation?
corr_pear = mvs.corr(method = 'pearson')
##high correlation
sns.heatmap(corr_pear,annot=True)
plt.title('Correlation Matrix for Movie Attributes')
plt.xlabel('Movie Attributes')
plt.ylabel('Movie Attributes')
```
